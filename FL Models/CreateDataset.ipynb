#Creating Dataset for IT Services and Products

**Services are defined based on 10 parameters :**

- Multi-tenancy model   (Numerical : (1 to 1000)
- Automated provisioning  (Categorical : Low/Medium/High)
- Single Sign On   (Categorical : Yes/No)
- Subscription based billing   (Categorical : Low/Medium/High)
- High availability  (Categorical : Low/Medium/High)
- Elastic Infrastructure   (Categorical : Strongly Disagree/Disagree/Neutral/Agree/Strongly Disagree)
- Data Security   (Categorical : Low/Medium/High)
- Application Security  (Categorical : Low/Medium/High)
- Rate limiting/QoS  (Numerical : (1 to 100 requests per second)
- Audit   (Categorical : Low/Medium/High)

**A neural networks learns on these 10 parameters using regression/classification modelling to predict following 4 features :**

- Security
- Scalability
- Ease Of USe
- Latency

```python
# Loading Libraries

import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
# Function for visualizing the distribution of data

def visualize(values, title, xlabel, ylabel):
  plt.hist(values)
  plt.title(title)
  plt.xlabel(xlabel)
  plt.ylabel(ylabel)
```

```python
# Defining the training and prediction features

numRows = 10000

inputFeatureList = ["Multi-Tenancy model", "Automated Provisioning", "Single Sign On", "Subscription Based Billing", "High Availability", "Elastic Infrastructure",	"Data Security", "Application Security", "Rate Limiting/QoS", "Audit"]				
outputInferenceList = ["Security",	"Scalability",	"Ease Of Use",	"Latency"]

print("Number of Input features  :", len(inputFeatureList))
print("Input Features : ", inputFeatureList)
print("Number of Output Inference Features :", len(outputInferenceList))
print("Output Features : ", outputInferenceList)
```

# Generating training features

- Multi-tenancy model   (Numerical : (1 to 1000)
- Automated provisioning  (Categorical : Low/Medium/High)
- Single Sign On   (Categorical : Yes/No)

- High availability  (Categorical : Low/Medium/High)
- Elastic Infrastructure   (Categorical : Strongly Disagree/Disagree/Neutral/Agree/Strongly Disagree)

```python
# Keeping one field of service ID as primary key

services = []

for i in range(numRows):
    services.append("Service_ID : " + str(i+1))
```

```python
multiTenancy = pd.DataFrame(multiTenancy)
automatedProvisioning = pd.DataFrame(automatedProvisioning)
sso = pd.DataFrame(sso)
subBilling = pd.DataFrame(subBilling)
avail = pd.DataFrame(avail)
elastic = pd.DataFrame(elastic)
dataSecurity = pd.DataFrame(dataSecurity)
appSecurity = pd.DataFrame(appSecurity)
qos = pd.DataFrame(qos)
audit = pd.DataFrame(audit)
services = pd.DataFrame(services)

inputFeatures = pd.concat([services, multiTenancy, automatedProvisioning, sso, subBilling, avail, elastic, dataSecurity, appSecurity, qos, audit], axis=1)

inputFeatures.columns = ["Service Id"] + inputFeatureList

inputFeatures.head()
```

# Generating prediction features

- Security
- Scalability
- Ease Of USe
- Latency

```python
# Security

outSecure = []

mp = {"Low":1, "Medium":2, "High":3}

for i in range(numRows):

    if mp[inputFeatures["Data Security"][i]] * mp[inputFeatures["Application Security"][i]] >=6:
        outSecure.append("High")

    elif mp[inputFeatures["Data Security"][i]] * mp[inputFeatures["Application Security"][i]] >=4:
        outSecure.append("Medium")

    else:
        outSecure.append("Low")

visualize(outSecure, "Security rating", "Rating", "Number of SaaS")
print(outSecure)
```

```python
# Scalability

outScale = []

for i in range(numRows):

    if inputFeatures["Multi-Tenancy model"][i] * inputFeatures["Rate Limiting/QoS"][i] >=60*600:
        outScale.append("High")

    elif inputFeatures["Multi-Tenancy model"][i] * inputFeatures["Rate Limiting/QoS"][i] >=40*400:
        outScale.append("Medium")

    else:
        outScale.append("Low")

visualize(outScale, "Scalability rating", "Rating", "Number of SaaS")
print(outScale)
```

```python
# Ease of Use

ease = []

mp1 = {"Low":1, "Medium":2, "High":3}
mp2 = {"Strongly Disagree":1, "Disagree":2, "Neutral":3, "Agree":4, "Strongly Agree":5}



for i in range(numRows):


    if mp1[inputFeatures["Automated Provisioning"][i]] * mp1[inputFeatures["Subscription Based Billing"][i]] * mp1[inputFeatures["High Availability"][i]] * mp2[inputFeatures["Elastic Infrastructure"][i]] * mp1[inputFeatures["Audit"][i]] >=2*2*2*2*4:
        
        if inputFeatures["Single Sign On"][i]=="Yes":
            ease.append("Strongly Agree")
        else:
            ease.append("Agree")

    elif mp1[inputFeatures["Automated Provisioning"][i]] * mp1[inputFeatures["Subscription Based Billing"][i]] * mp1[inputFeatures["High Availability"][i]] * mp2[inputFeatures["Elastic Infrastructure"][i]] * mp1[inputFeatures["Audit"][i]] >=2*2*1*2*3:
        
        if inputFeatures["Single Sign On"][i]=="Yes":
            ease.append("Agree")
        else:
            ease.append("Neutral")

    else:
        if inputFeatures["Single Sign On"][i]=="Yes":
            ease.append("Disagree")
        else:
            ease.append("Strongly Disagree")

visualize(ease, "Ease of use rating", "Rating", "Number of SaaS")
print(ease)
```

```python
# Latency

latency = []

mp1 = {"Low":1, "Medium":2, "High":3}
mp2 = {"Strongly Disagree":1, "Disagree":2, "Neutral":3, "Agree":4, "Strongly Agree":5}



for i in range(numRows):


    if mp1[inputFeatures["Automated Provisioning"][i]] * mp1[inputFeatures["Subscription Based Billing"][i]] * mp1[inputFeatures["High Availability"][i]] * mp2[inputFeatures["Elastic Infrastructure"][i]] * (inputFeatures["Rate Limiting/QoS"][i]/20) >=2*2*2*4*4:
        
        if inputFeatures["Single Sign On"][i]=="Yes":
            latency.append("Very Low")
        else:
            latency.append("Low")

    elif mp1[inputFeatures["Automated Provisioning"][i]] * mp1[inputFeatures["Subscription Based Billing"][i]] * mp1[inputFeatures["High Availability"][i]] * mp2[inputFeatures["Elastic Infrastructure"][i]] * (inputFeatures["Rate Limiting/QoS"][i]/20) >=2*2*1*2*2:
        
        if inputFeatures["Single Sign On"][i]=="Yes":
            latency.append("Low")
        else:
            latency.append("Medium")

    else:
        if inputFeatures["Single Sign On"][i]=="Yes":
            latency.append("High")
        else:
            latency.append("Very High")

visualize(latency, "Latency rating", "Rating", "Number of SaaS")
print(latency)
```

```python
outSecure = pd.DataFrame(outSecure)
outScale = pd.DataFrame(outScale)
ease = pd.DataFrame(ease)
latency = pd.DataFrame(latency)

outputFeatures = pd.concat([outSecure, outScale, ease, latency], axis=1)

outputFeatures.columns = outputInferenceList

outputFeatures.head()
```

# Generating a final score ( out of 100 ) based on weightage from each of 4 predicted features

```python
# Final grading score

alpha, beta, gamma, delta = 5, 3, 4, 4     # Weightage of each parameter in final scoring

max_score = alpha*3 + beta*3 + gamma*5 + delta*5

mp1 = {"Low":1, "Medium":2, "High":3}
mp2 = {"Strongly Disagree":1, "Disagree":2, "Neutral":3, "Agree":4, "Strongly Agree":5}
mp3 = {"Very High":1, "High":2, "Medium":3, "Low":4, "Very Low":5}
finalScore = []

for i in range(outputFeatures.shape[0]):
    val_1 = mp1[outputFeatures["Security"][i]]*alpha
    val_2 = mp1[outputFeatures["Scalability"][i]]*beta
    val_3 = mp2[outputFeatures["Ease Of Use"][i]]*gamma
    val_4 = mp3[outputFeatures["Latency"][i]]*delta

    finalScore.append((val_1 + val_2 + val_3 + val_4)*100/max_score)

visualize(finalScore, "Final Grading", "Scores", "Number of SaaS")
finalScore = pd.DataFrame(finalScore)
finalScore.columns = ["Score"]
#finalScore.max()
#finalScore.head()
```

```python
finalDataset = pd.concat([inputFeatures, outputFeatures, finalScore], axis=1)
finalDataset
```
